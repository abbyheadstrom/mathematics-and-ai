```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 10)
        print(' ')

def check_winner(board, mark):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],
    ]
    return [mark, mark, mark] in win_conditions

def check_if_empty(board, entry): 
    empty_positions = get_empty_positions(board)
    return entry in empty_positions

def get_empty_positions(board):
    positions = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                positions.append((i, j))
    return positions

def user_move(board):
    while True:
        try:
            row = int(input("Enter the row (1-3): ")) - 1
            col = int(input("Enter the column (1-3): ")) - 1
            if board[row][col] == " ":
                board[row][col] = "X"
                break
            else:
                print("This position is already taken.")
        except (ValueError, IndexError):
            print("Invalid input. Please enter numbers between 1 and 3.")
            
def detect_next_winning_move(board, mark):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],
    ]
    
    for i in win_conditions:
        if i == [mark, " ", mark]:
            return True
        elif i == [" ", mark, mark]:
            return True
        elif i == [mark, mark, " "]:
            return True
        
    return False

def return_winning_move(board, mark):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],
    ]
    
    i = 0
    while i < len(win_conditions):
        if win_conditions[i] == [" ", mark, mark]:
            if i == (0 or 1 or 2):
                return [i,0]
            elif i == 3:
                return [0,0]
            elif i == 4: 
                return [0,1]
            elif i == 5: 
                return [0,2]
            elif i == 6:
                return [0,0]
            elif i == 7:
                return [2,0]
        elif win_conditions[i] == [mark, " ", mark]:
            if i == (0 or 1 or 2):
                return [i,1]
            elif i == 3:
                return [1,0]
            elif i == 4: 
                return [1,1]
            elif i == 5: 
                return [1,2]
            elif i == 6:
                return [1,1]
            elif i == 7:
                return [1,1]
        elif win_conditions[i] == [mark, mark, " "]:
            if i == (0 or 1 or 2):
                return [i,2]
            elif i == 3:
                return [2,0]
            elif i == 4: 
                return [2,1]
            elif i == 5: 
                return [2,2]
            elif i == 6:
                return [2,2]
            elif i == 7:
                return [0,2]
            
        i +=1
            

def computer_move(board, move_num):
    empty_positions = get_empty_positions(board)
    # this is the strategy if the computer goes first 
    if move_num % 2 == 1:
        if move_num == 1:
            board[0][0]= "O"
            
        # for the third move of the game, we either play in the bottom right corner or the bottom left if that corner is taken
        elif move_num ==3:
            i = (2,2)
            if not check_if_empty(board, i):
                board [2][0]= "O"
            else:
                board[2][2] = "O"
                
        # for move # 5, we first have to check if there is a possibility for either the computer or the other player to win. If yes, then we either win the game or block them.
        elif move_num == 5:
            if detect_next_winning_move(board, "O"):
                move = return_winning_move(board, "O")
                row = move[0]
                col = move[1]
                board[row][col] = "O"
                
            elif detect_next_winning_move(board, "X"):
                move = return_winning_move(board, "X")
                row = move[0]
                col = move[1]
                board[row][col] = "O"    
              
            # note that there is only one possible scenario in which X or O does not have a winning move on move 5/6. In this case, we put the X in the top right corner  
            else:
                board[0][2] = "O"
        # at move #7, every position the computer can play is either winning or stopping the other player from winning   
        elif move_num == 7:
            if detect_next_winning_move(board, "O"):
                move = return_winning_move(board, "O")
                row = move[0]
                col = move[1]
                board[row][col] = "O"
                
            elif detect_next_winning_move(board, "X"):
                move = return_winning_move(board, "X")
                row = move[0]
                col = move[1]
                board[row][col] = "O" 
                
        # the last play there will be only one position left
        else: 
            move = empty_positions[0]
            row = move[0]
            col = move[1]
            board[row][col] = "O"
                
            
    # this is the strategy if the computer goes second 
    elif move_num % 2 == 0: 
        # first check if there is either a way to win the game or stop the other player from winning
        if detect_next_winning_move(board, "O"):
                move = return_winning_move(board, "O")
                row = move[0]
                col = move[1]
                board[row][col] = "O"
                
        elif detect_next_winning_move(board, "X"):
            move = return_winning_move(board, "X")
            row = move[0]
            col = move[1]
            board[row][col] = "O" 
        
        # if not, then the computer chooses a spot at random
        else: 
            
            move = random.choice(empty_positions)
            board[move[0]][move[1]] = "O"
    
        
        
    
    

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic Tac Toe!")
    user_first = input("Do you want to go first? (y/n): ").lower() == 'y'
    
    move_num = 0
    for _ in range(9):
        move_num += 1
        print_board(board)
        if user_first:
            user_move(board)
            if check_winner(board, "X"):
                print_board(board)
                print("Congratulations! You win!")
                return
            user_first = False
        else:
            computer_move(board, move_num)
            if check_winner(board, "O"):
                print_board(board)
                print("Computer wins! Better luck next time.")
                return
            user_first = True
        
        if not get_empty_positions(board):
            print_board(board)
            print("It's a draw!")
            return

if __name__ == "__main__":
    tic_tac_toe()
```
